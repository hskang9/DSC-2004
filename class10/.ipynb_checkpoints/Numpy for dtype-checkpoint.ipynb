```python
import pandas as pd
import numpy as np

# Use pandas to read csv file
ct_cp = pd.read_csv('countrycapitalsmodified.csv',error_bad_lines=False)
print(ct_cp.to_string())

# Change it into 2 dimension numpy array
ct_cp_np = ct_cp.as_matrix()
print(ct_cp_np)

# Wrap Each array for stacking up
ct_cp_np = np.array(ct_cp_np)
row = np.array([['Korea', 'Seoul', 37.5665, 126.9780, 'KR', 'Asia']])

# Stack vertically
np.vstack((ct_cp_np, row))

# Print Result
print(ct_cp.to_string())

# Print elements whose ContinentName column has Asia
print(ct_cp.loc[ct_cp["ContinentName"]=="Asia"])
```
