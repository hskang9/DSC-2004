# 1. Electricity Production using stacked plot chart


```python
import pandas as pd
import numpy as np
```

```python
data = {}
d = pd.read_csv("Energy-Consumption.csv", encoding='utf-8') # get data from file stream
data["Energy-Consumption.csv".replace(".csv", "")] = d # put data in each category and make dictionary(ap classes on 2010, class size, demographics of the schools, graduation, high school directory, math test results, sat results)
    
```

```python

months = data["Energy-Consumption"].iloc[:, 2:]

data["Energy-Consumption"]["Country"]
```

```python
import matplotlib.pyplot as plt
x = np.arange(1,13)
y = {}
np_months = months.as_matrix()
```

```python
for idx, consumption in enumerate(np_months):
        y[data["Energy-Consumption"]["Country"][idx]] = consumption
fig, ax = plt.subplots()

consumption = []
for idx, k in enumerate(y.keys()):
    consumption.append(y[k])
plt.tight_layout()
ax.stackplot(x,consumption)
ax.legend(y.keys())

    
plt.show()
```

# 2. Precipitation using pie chart

```python
import matplotlib.pyplot as plt

frac=[20,21.3,21.3,37.3]
labels=['Spring', 'Summer', 'Autumn', 'Winter']

fig = plt.figure(1, figsize=(6,6))
ax = fig.add_subplot(111)
ax.set_title('Rainy days by season')
ax.axis('equal')
explode = (0.1, 0.1, 0.1, 0.1) 
colors=('b', 'g', 'r', 'c')
ax.pie(frac,colors=colors ,labels=labels, autopct='%1.1f%%', explode=explode)
plt.show()
```

# 3. Sales Net Report using 3D Projection

```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
```

```python
mpl.rcParams['font.size'] = 9
fig = plt.figure()
ax = fig.add_subplot(111, animated=False, projection='3d')
```

```python

yn = list(y.keys())[:4]
for c, z, key in zip(['r', 'g', 'b', 'y'], [30, 20, 10, 0], yn):
    xs = np.arange(1,13)
    ys = y[key]
    cs = [c] * (len(xs)+1)
    ax.bar(xs, ys, zs=z,zdir='y', color=cs, alpha=0.8)
    
ax.set_xlabel("Month")
ax.set_ylabel("Countries")
ax.set_zlabel("Consumption")


plt.show()
```
